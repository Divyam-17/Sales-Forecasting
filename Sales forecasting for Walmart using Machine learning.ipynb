```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from scipy.special import boxcox1p
import seaborn as sns
```

```python
#Getting data
features=pd.read_csv(r"C:\Users\divya\OneDrive\Desktop\Walmart Dataset\features.csv")
store=pd.read_csv(r"C:\Users\divya\OneDrive\Desktop\Walmart Dataset\stores.csv")
train=pd.read_csv(r"C:\Users\divya\OneDrive\Desktop\Walmart Dataset\train.csv")
test=pd.read_csv(r"C:\Users\divya\OneDrive\Desktop\Walmart Dataset\test.csv")
```

```python
#reseting train
train=train.groupby(['Store','Date'])['Weekly_Sales'].sum()
train=train.reset_index()
train.head(10)
```

```python
#merging train and feature
data=pd.merge(train,features,on=['Store','Date'],how='inner')
data.head(10)
```

```python
#marging store with data
data=pd.merge(data,store,on=['Store'],how='inner')
data.head(10)
```

```python
#sorting values of Data
data=data.sort_values(by='Date')
data.head(10)
```

```python
data['Holiday']=[int(i) for i in list(data.IsHoliday)]
data.head(10)
```

```python
Type_dummy=pd.get_dummies(data['Type'],drop_first=True)
Type_dummy.head(10)
```

```python
data=pd.concat([data,Type_dummy],axis=1)
data.head(10)
```

```python
data=data.drop(['Type','IsHoliday'],axis=1)
data.drop(10)
```

```python
#splitting data in input and output
X=data.drop(['Weekly_Sales','Store','Date'],axis=1)
y=data['Weekly_Sales']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
LR=LinearRegression(normalize=True)
LR.fit(X_train,y_train)
```

```python
y_pred=LR.predict(X_test)
```

```python
Root_mean_square_error=np.sqrt(np.mean(np.square(y_test-y_pred)))
print(Root_mean_square_error)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print(r2)
```
